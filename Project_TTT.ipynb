```python
from IPython.display import clear_output
import random

def display_board(board):
    clear_output()  
    
    print('         |                    |           ')
    print(board[7]+'        |          '+board[8]+'         |          '+board[9])
    print('_________|____________________|____________')
    print('         |                    |           ')
    print(board[4]+'        |          '+board[5]+'         |          '+board[6])
    print('_________|____________________|____________')
    print('         |                    |           ')
    print(board[1]+'        |          '+board[2]+'         |          '+board[3])


    
def player_input():
    marker = ''
    
    while not (marker == 'X' or marker == 'O'):
        marker = input('Player 1 Master : Do you want to be X or O? ').upper()

    if marker == 'X':
        return ('X', 'O')
    else:
        return ('O', 'X')         #player1_marker, player2_marker =  player_input()
    
    
    
    
def place_marker(board, marker, position):
    board[position] = marker
    
    
    

def win_check(board,mark):
    
    if board[1] == board [2] == board[3] == mark:
        return (True)
    elif board[4] == board [5] == board[6] == mark:
        return (True)
    elif board[7] == board [8] == board[9] == mark:
        return (True)
    elif board[7] == board [4] == board[1] == mark:
        return (True)
    elif board[8] == board [5] == board[2] == mark:
        return (True)
    elif board[9] == board [6] == board[3] == mark:
        return (True)
    elif board[7] == board [5] == board[3] == mark:
        return (True)
    elif board[9] == board [5] == board[1] == mark:
        return (True)
    else :
        return (False)





def choose_first():
    r = random.randint(0,10)
    if r == 1 or r == 2 or r == 3 or r == 4 or r == 5:
        return ('Player 1')
    else :
        return ('Player 2')
    
    
    
def space_check(board, position):
    
    return board[position] == ' '



def full_board_check(board):
    for i in range(1,10):
        if space_check(board, i):
            return (False)
    return (True)




def player_choice(board):
    position = 0
    
    while position not in range(1,10) or not space_check(board, position):
        position = int(input('Use Numpad and Choose your next position: (1-9) '))
        
    return (position)




def replay():
    
    r = input('Would you like to take your revenge!! or exit??, Press y for yes and n for no: ')
    if r == 'y' or r == 'Y':
        return (True)
    else :
        return (False)


print('Welcome to the TIC-TAC-TOE BATTLE !')

while True:
    # Reset the board
    theBoard = [' '] * 10
    
    #taking input for player1 & player2
    player1_marker, player2_marker = player_input()
    
    
    print('\nùôÇùôäùòøùôé will declare who will go first\n')
    
    
    #which player will go first
    chance = choose_first()
    print(chance + ' Master has been chosen worthy to go ùôÅùôÑùôçùôéùôè.')
    
    #checking if players are ready and input for next while loop run
    play_game = input('Are you ready to Slay the Enemy?? Press y for yes and n for no.')
    
    if play_game == 'y' or play_game == 'Y' :
        game_on = True
    else:
        game_on = False

    #game loop    
    while game_on == True:
        if chance == 'Player 1':
            # Player1's turn.
            
            #displaying the board
            display_board(theBoard)
            
            #checking if the input position is valid 
            position = player_choice(theBoard)
            
            #marking the position
            place_marker(theBoard, player1_marker, position)

            #checking conditions for winning, board_full and swapping chance
            if win_check(theBoard, player1_marker) == True:
                display_board(theBoard)
                print('Player 1 Master is Victorous !')
                game_on = False
            else:
                if full_board_check(theBoard) == True:
                    display_board(theBoard)
                    print('Both Masters are of equal Strength!')
                    game_on = False
                else:
                    chance = 'Player 2'

        else:
            # Player2's turn.
            
            display_board(theBoard)
            position = player_choice(theBoard)
            place_marker(theBoard, player2_marker, position)

            if win_check(theBoard, player2_marker) == True:
                display_board(theBoard)
                print('Player 2 Master is Victorous')
                game_on = False
            else:
                if full_board_check(theBoard) == True:
                    display_board(theBoard)
                    print('Both Masters are of equal Strength')
                    game_on = False
                else:
                    chance = 'Player 1'
    
    #breaking out of loop if don't want to play
    if replay() == False :
        break
        
        
        
        








```
